## PCA분석 
: 변수 두개가 선형연관성이 있을 경우, 이를 가로지르는 제 3 데이터(주성분) 생성    

서로 직각인 주성분끼리 연관성은 x(하나의 주성분의 길이나 너비가 변하더라도 다른 주성분은 변화가 없어야함)  

1. 분산이 최대인 첫 주성분 찾기 
2. 첫번째 주성분이 직교하는 두번째 주성분
- 또 다른 주성분: 또 다른 변수간의 선형 연관성 찾기
- 직교하는 주성분 = 서로 독립적인 주성분

```python
from sklearn.preprocessing import StandardScaler
iris_scaler = StandardScaler().fit(df_iris[feature_names])# iris_scaler:평균, 표편있는 인스턴스
iris_scaler_result = iris_scaler.transform(df_iris[feature_names])# 변환해달라
```

```python
df_iris_scaler = pd.DataFrame(iris_scaler_result, columns = feature_names)
df_iris_scaler['target'] = df_iris['target']

sns.pairplot(df_iris_scaler, diag_kind = 'kde', hue = 'target')
plt.show()
# 모든 속성이 동일한 범위내로 변경
```

```python
from sklearn.decomposition import PCA

iris_scale_pca = PCA(2) # 2개의 주성분으로 분석하겠다

# 표준점수로 변환한 dataframe
iris_pca_trans = iris_scale_pca.fit_transform(df_iris_scaler[feature_names])

print(iris_pca_trans.shape)
# 두 주성분을 df_iris에 pca_x, pca_y로 저장
df_iris['pca_x'] = iris_pca_trans[:,0]
df_iris['pca_y'] = iris_pca_trans[:,1]
```

```python
# 분류 결과인 열 target과 PCA 분석 결과를 매칭하여 산점도 생성(분류하기)
fig, ax = plt.subplots()
markers = ['^', 's','o']
for i in range(3):
    x_axis_data = df_iris[df_iris['target'] == i]['pca_x']
    y_axis_data = df_iris[df_iris['target'] == i]['pca_y']
    ax.scatter(x_axis_data, y_axis_data, marker = markers[i], label = iris_dataset.get('target_names')[i])
ax.legend()
plt.show()
```

```python
print(iris_scale_pca.explained_variance_ratio_)
# 각 주성분이 차지하는 비율의 합이 95퍼를 넘었을 때, 분석에 필요한 주성분 개수가 충분하다고 판단
```

# k-means

```python
from sklearn.cluster import KMeans
k_mean = KMeans(n_clusters = 3, init = 'k-means++') # 생성할 군집 개수
# init: 군집화 알고리즘
k_mean.fit(df_iris[feature_names])
```

```python
df_iris['kmean_cluster'] = k_mean.labels_
df_iris
```

```python
df_iris.groupby(['target','kmean_cluster'])['sepal_length'].count()
df_iris.groupby(['target','kmean_cluster'])['sepal_width'].count()
df_iris.groupby(['target','kmean_cluster'])['petal_length'].count()
df_iris.groupby(['target','kmean_cluster'])['petal_width'].count()
```

```python
fig, axes = plt.subplots(2,1)
markers = ['^','s','o']
markers_k = ['o','^','s']
for i in range(3):
    x_axis_data = df_iris[df_iris['target']==i]['pca_x']
    y_axis_data = df_iris[df_iris['target']==i]['pca_y']
    axes[0].scatter(x_axis_data, y_axis_data, marker=markers[i], label = iris_dataset.get('target_names')[i])
for i in range(3):
    x_axis_data = df_iris[df_iris['kmean_cluster']==i]['pca_x']
    y_axis_data = df_iris[df_iris['kmean_cluster']==i]['pca_y']
    axes[1].scatter(x_axis_data, y_axis_data, marker=markers_k[i], label = iris_dataset.get('target_names')[i])
axes[0].legend()
axes[1].legend()
plt.show()
```

```python
k_mean.cluster_centers_
```

```python
# 표준 점수로 변환
iris_scaler_center = iris_scaler.transform(pd.DataFrame
                                           (k_mean.cluster_centers_,
                                            columns = feature_names))
k_mean_center_x =  iris_scale_pca.transform(pd.DataFrame
                                         (iris_scaler_center,
                                          columns = feature_names))[:,:1] #n행 1열
k_mean_center_y = iris_scale_pca.transform(pd.DataFrame(iris_scaler_center,columns = feature_names))[:,1:2]
k_mean_center_y
# 행렬 슬라이싱 조심!
```

```python
fig, ax = plt.subplots()
markers = ['^','s','o']
for i in range(3):
    x_axis_data = df_iris[df_iris['kmean_cluster']==i]['pca_x']
    y_axis_data = df_iris[df_iris['kmean_cluster']==i]['pca_y']
    ax.scatter(x_axis_data, y_axis_data, marker=markers[i], label = iris_dataset.get('target_names')[i])
ax.scatter(k_mean_center_x, k_mean_center_y, c='r')
ax.legend()
ax.legend()
plt.show()
```

```python
# 엘보우 기법
elbow_inertias = list()
for i in range(1,10):
    k_mean_elbow  = KMeans(n_clusters=i, init='random', n_init = 15, max_iter = 150, random_state = 0)
    k_mean_elbow.fit(df_iris[feature_names])
    elbow_inertias.append(k_mean_elbow.inertia_)
plt.plot(range(1,10), elbow_inertias, marker = 'o')
plt.xlabel('Number of Cluster')
plt.ylabel('Sum of squared distances')
plt.show()
```
