```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)
size = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')
for row in data:
    if name in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m += int(row[i+3])
            f += int(row[i+106])
        break
size.append(m)
size.append(f)
print(size)
plt.rc('font', family = 'Malgun Gothic')
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels = ['남','여'], autopct='%.1f%%', colors = color, startangle= 90)
plt.title(name + ' 지역의 남녀 성별 비율')
plt.show()
```

```python
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')
for row in data:
    if name in row[0]:
        for i in range(3, 101):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
        break
plt.plot(m, label = 'Male')
plt.plot(f, label = 'Female' )
plt.legend()
plt.show()
```

```python
import numpy as np
f = open('gender.csv')
data = csv.reader(f)
result = []
name = input('궁금한 동네를 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            result.append(int(row[i]) - int(row[i+103]))
        break
plt.bar(np.arange(0,101),result)
plt.show()
```

```python
f = open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
max_s = 0
for row in data:
    row[4:] = map(int, row[4:])
    tmp = row[10] + row[12]+row[14]
    if tmp > max_s:
        result = []
        result.append(row[3])
        result.append(tmp)
        max_s = tmp
print(result[0], result[1])
```

```python
f = open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)
max_s = 0
time = input('몇 시의 승차 인원이 가장 많은 역이 궁금하세요? :')
t = 4 + (int(time)-4)*2
for row in data:
    row[4:] = map(int, row[4:])
    if row[t] > max_s:
        result = []
        result.append(row[3])
        result.append(row[t])
        max_s = row[t]
print(result[0], result[1])
```
