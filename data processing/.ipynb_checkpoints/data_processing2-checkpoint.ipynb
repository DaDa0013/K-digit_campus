```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            result.append(float(row[-1]))

plt.hist(result, bins = 100, color ='r')
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
            result.append(float(row[-1]))

plt.plot(result, 'hotpink')
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = []
low = []

for row in data:
    if row[-1] != '':
        if int(row[0].split('-')[0]) >= 1983:
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
            
plt.plot(high, 'hotpink')
plt.plot(low, 'blue')
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result1 = []
result8 = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            result8.append(float(row[-1]))
        if row[0].split('-')[1] == '01':
            result1.append(float(row[-1]))

plt.hist(result8, bins = 100, color ='r')
plt.hist(result1, bins = 100, color = 'b')
plt.show()
```

```python
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
for row in data:
    if '신도림' in row[0]:
        for i in range(0,101):
            m.append(int(row[i+3]))
            f.append(int(row[-(i+1)]))
f.reverse()
```

```python
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
for row in data:
    if '신도림' in row[0]:
        for i in row[3:104]:
            m.append(-int(i))
        for i in row[106:]:
            f.append(int(i))
```

```python
plt.rc('font', family= 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

```python
name = input('찾고 싶은 지역의 이름을 알려주세요 :')
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m.append(-int(i))
        for i in row[106:]:
            f.append(int(i))
plt.rc('font', family= 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + ' 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
```
