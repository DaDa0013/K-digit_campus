```python
import matplotlib.pyplot as plt

f = open('subwayfee.csv')
data = csv.reader(f)
result_seoul = []
result_s = []
next(data)
for row in data:
    if row[3]+' '+row[1] == '서울역 1호선':
        for i in range(4,8):
            result_seoul.append(int(row[i]))
    if row[3]+' '+row[1] == '시청 1호선':
        for i in range(4,8):
            result_s.append(int(row[i]))
plt.rc('font', family= 'Malgun Gothic')
plt.subplot(2, 1, 1)
plt.pie(result_seoul)
plt.title('서울역 1호선')
plt.subplot(2, 1, 2)
plt.pie(result_s)
plt.title('시청 1호선')
plt.show()
```

```python
import matplotlib.pyplot as plt

f = open('subwayfee.csv')
data = csv.reader(f)
result_seoul = []
result_s = []
c = ['lightgreen', 'green', 'hotpink', 'purple']
label = ['유임승차','유임 하차','무임승차','무임하차']
next(data)
for row in data:
    if row[3]+' '+row[1] == '서울역 1호선':
        for i in range(4,8):
            result_seoul.append(int(row[i]))
    if row[3]+' '+row[1] == '시청 1호선':
        for i in range(4,8):
            result_s.append(int(row[i]))
plt.rc('font', family= 'Malgun Gothic')
plt.subplot(2, 1, 1)
plt.pie(result_seoul, labels = label, colors = c, autopct = '%1.f%%')
plt.title('서울역 1호선')
plt.subplot(2, 1, 2)
plt.pie(result_s, labels = label, colors = c,autopct = '%1.f%%')
plt.title('시청 1호선')
plt.show()
```

```python
import matplotlib.pyplot as plt
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)
name = input('출력할 호선을 입력하세요 : ')
y = []
x = []
for row in data:
    if row[1] == name:
        y.append(int(row[4]))
        x.append(row[3])
plt.rc('font', family= 'Malgun Gothic')
plt.bar(x,y)
plt.style.use('ggplot')
plt.ticklabel_format(axis = 'y', useOffset = False,style = 'plain')
plt.xticks(rotation=90)
plt.title(name)
plt.show()
```

```python
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)
name = input('출력할 호선과 역을 입력하세요 : ')
label = ['유임승차','유임 하차','무임승차','무임하차']
result = []
for row in data:
    if row[1] == name.split()[0] and row[3] == name.split()[1]:
        for i in range(4,8):
            result.append(int(row[i]))
        break

plt.rc('font', family= 'Malgun Gothic')
plt.bar(label,result, color = 'blue')
plt.gca().set_facecolor('white') #배경색
plt.ticklabel_format(axis = 'y', useOffset = False,style = 'plain')
plt.title(name)
plt.show()
```
