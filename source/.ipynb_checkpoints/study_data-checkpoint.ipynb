# 데이터의 이해

## pandas

```python
import pandas as pd

data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data'
df_data = pd.read_csv(data_url, sep='\s+', header = None) # header 없이 불러오기
df_data.columns = ['CRIM', 'ZN', 'INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PRTATIO', 'B', 'LSTAT','MEDV'] # 데이터 열 이름
df_data.head()
```

```python
import numpy as np
df_data['weight_0'] = 1 # column 추가
df_data= df_data.drop("MEDV", axis = 1) # y값 제거
df_data.head()
```

```python
df_data.values
```

```python
df_matrix = df_data.values# 행렬 데이터로 변환
```

```python
weight_vector = np.random.random_sample((14,1)) # 가중치
df_matrix.dot(weight_vector) # 내적연산
```
