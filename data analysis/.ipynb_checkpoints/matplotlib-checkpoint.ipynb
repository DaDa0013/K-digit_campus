```python
randn_cumsum = np.random.randn(1000).cumsum() # 
fig, ax = plt.subplots()

ax.set_title('my first matplotlib plot')
ax.set_xlabel('Stages')
ax.set_ylabel('Random Value')
ax.set_xticks([0, 250, 500, 750, 1000])
ax.set_xticklabels(['one','two', 'three', 'four', 'five'], rotation=30, fontsize = 'small')
ax.annotate('250\n250.00\n', xy = (250, randn_cumsum[250] + 0.5), xytext = (250, randn_cumsum[250]+10), arrowprops = dict(facecolor ='black', headwidth = 4, width = 2, headlength = 4),horizontalalignment = 'right', verticalalignment = 'top')
ax.plot(randn_cumsum)
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

np_randn = np.random.randn(50)

ax1.hist(np_randn, bins = 20, color= 'k', alpha = 0.3)
ax2.scatter(np.arange(50), np.arange(50)+3*np_randn)
ax3.plot(np_randn.cumsum(), 'k--')

plt.show()
```

```python
np_randn = np.random.randn(50)

fig,axes = plt.subplots(2,3)

axes[0,0].hist(np_randn, bins = 20, color = 'k', alpha  = 0.3)
axes[0,2].scatter(np.arange(50), np.arange(50) +3*np_randn)
axes[1,1].plot(np_randn.cumsum(), 'k--')
plt.show()
```

```python
fig, axes = plt.subplots(2,2)

for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500), bins = 50, color = 'k', alpha = 0.5)
fig.subplots_adjust(wspace = 0, hspace = 0) #,  
plt.show()
```

```python
plt.plot(randn_cumsum, 'ko--')
plt.plot(randn_cumsum, linestyle='--', color = 'k', marker = 'o')
plt.plot(randn_cumsum, 'ko--', np_randn, 'r-')
plt.show()
```

```python
randn1_cumsum = np.random.randn(500).cumsum()
randn2_cumsum = np.random.randn(500).cumsum()
randn3_cumsum = np.random.randn(500).cumsum()

fig, axes = plt.subplots(2,2)

axes[0,0].set_title('randn1_cumsum')
axes[0,1].set_title('randn2_cumsum')
axes[1,0].set_title('randn3_cumsum')
axes[1,1].set_title('randn_all_cumsum')

axes[0,0].plot(randn1_cumsum, 'k--')
axes[0,1].plot(randn2_cumsum, 'k--')
axes[1,0].plot(randn3_cumsum, 'k--')
axes[1,1].plot(randn1_cumsum, 'r--', label = 'randn1_cumsum')
axes[1,1].plot(randn2_cumsum, 'b--', label = 'randn2_cumsum')
axes[1,1].plot(randn3_cumsum, 'go-', label = 'randn3_cumsum')
axes[0,0].legend()
axes[1,1].legend(loc='best')
ax.plot(randn_cumsum)
plt.show()
```

```python
np_randn = np.random.randn(7)
label_x = list()
for i in range(7):
    label_x.append(str(i))
fig,axes = plt.subplots(2,1)

axes[0].set_xticks(range(7))
axes[0].set_xticklabels(label_x)
axes[0].bar(range(7), np_randn, color='k', alpha = 0.7)

axes[1].set_yticks(range(7))
axes[1].set_yticklabels(label_x)
axes[1].barh(range(7), np_randn, color='k', alpha = 0.7)
plt.show()
```

```python
ss = pd.Series(np.random.randn(10).cumsum(), index = np.arange(0,100,10))
ax = ss.plot()
plt.show()
```
